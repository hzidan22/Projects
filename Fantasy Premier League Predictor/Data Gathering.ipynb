```python
import requests
import pandas as pd
```

```python
url = "https://fantasy.premierleague.com/api/bootstrap-static/" 

response = requests.get(url)
if response.status_code == 200:
    data = response.json() 
    print("Data successfully retrieved!")
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
```

```python
players_df = pd.DataFrame(data["elements"])
players_df.head()
```

```python
teams_df = pd.DataFrame(data["teams"])
```

```python
teams_df
```

```python
url2 = "https://fantasy.premierleague.com/api/fixtures/"
response2 = requests.get(url2)
fixtures_data = response2.json()
fixtures_df = pd.DataFrame(fixtures_data)
fixtures_df.head()
```

```python
fixtures_df.to_csv("fixtures24-25.csv", index=False)  # Saves without index
```

```python
teams_df.to_csv("teams24-25.csv", index = False)
```

```python
columns_needed = [
    "first_name", "second_name", "goals_scored", "assists", "total_points", 
    "minutes", "goals_conceded", "creativity", "influence", "threat", 
    "bonus", "bps", "ict_index", "clean_sheets", "red_cards", 
    "yellow_cards", "selected_by_percent", "now_cost", "element_type", "team"
]
filtered_players_df = players_df[columns_needed]
print(filtered_players_df.tail(20))
```

```python
print(filtered_players_df["element_type"].dtype)
```

```python
element_type_mapping = {
    1: "GK",
    2: "DEF",
    3: "MID",
    4: "FWD", 
    5:"Manager"
}

filtered_players_df["element_type"] = filtered_players_df["element_type"].astype(int)
filtered_players_df.loc[:, "element_type"] = filtered_players_df["element_type"].map(element_type_mapping)

print(filtered_players_df.head())
```

```python
filtered_players_df_without = filtered_players_df.drop(['team'], axis = 1)
filtered_players_df_without.info()
```

```python
filtered_players_df.info()
```

```python
filtered_players_df['now_cost'] = filtered_players_df['now_cost'].astype(float)
filtered_players_df['now_cost'] = filtered_players_df['now_cost'] / 10
filtered_players_df.tail()
```

```python
filtered_players_df_without.to_csv("cleaned_players24-25.csv", index = False)
```
